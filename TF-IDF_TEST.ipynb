```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
from wordcloud import WordCloud

text = open('測試用/rfc1883.txt', 'r').read()
text1 = open('測試用/rfc1702.txt', 'r').read()
text2 = open('測試用/rfc2734.txt', 'r').read()
documents = [text,text1,text2]

vectorizer = TfidfVectorizer(stop_words='english')

tfidf = vectorizer.fit_transform(documents)
df_tfidf = pd.DataFrame(tfidf.toarray(),columns=vectorizer.get_feature_names(), index=['text','text1','text2'])

true_k = 3
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(tfidf)

print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]

terms = vectorizer.get_feature_names()

for i in range(true_k):
    textcloud = []
    print("Cluster %d:" % i)
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind])
        textcloud.append(terms[ind])
    print(textcloud)
    str6 = " ".join(textcloud)
    cloud = WordCloud().generate(str6)
    cloud.to_file('output%d.png' %i)
```
